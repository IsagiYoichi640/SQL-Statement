```python
import sqlite3
database = 'database.sqlite'
conn = sqlite3.connect(database)
print('Opened data successfully')
```

```python
import pandas as pd
teams = pd.read_sql("""SELECT * FROM TEAM;""",conn)
teams
```

```python
matches = pd.read_sql("""SELECT * FROM Match;""",conn)
matches
```

```python
DD = pd.read_sql("""SELECT * FROM Match WHERE Match_Winner == 6;""", conn)
DD
```

```python
DD_S8_S9 = pd.read_sql("""SELECT * FROM Match WHERE Match_Winner == 6 and Season_Id IN (8,9);""", conn)
DD_S8_S9
```

```python
result1=pd.read_sql("""SELECT AVG(Win_Margin),Match_Winner FROM Match
WHERE Season_Id==5
 GROUP BY Match_Winner ORDER BY AVG(Win_Margin);""",conn)
result1
```

```python
result2 = pd.read_sql("""SELECT COUNT(DISTINCT Venue_Id)FROM Match
WHERE Season_Id==2;""",conn)
result2
```

```python
result3 = pd.read_sql("""SELECT MIN(Win_Margin), MAX(Win_Margin),
               AVG(Win_Margin), COUNT(DISTINCT(Man_of_the_Match))
               FROM Match;""",conn)
result3
```
